# Cron Jobs Log Reader
Reads the first lines of all log files in `/var/log/cron-jobs`

```python
from pathlib import Path
from collections import deque

LOG_DIR = '/var/log/cron-jobs'
NUM_LINES = 10

def read_last_lines(filepath, n=NUM_LINES):
    """Read the last n lines of a file."""
    try:
        with open(filepath, 'r', encoding='utf-8', errors='replace') as f:
            lines = [line.rstrip() for line in deque(f, maxlen=n)]
    except PermissionError:
        lines = ['[Permission denied]']
    except Exception as e:
        lines = [f'[Error reading file: {e}]']
    return lines

def get_log_files(log_dir):
    """Return all files in the log directory (recursively)."""
    log_dir = Path(log_dir)
    if not log_dir.exists():
        print(f'Directory not found: {log_dir}')
        return []
    return sorted(log_dir.rglob('*'))

files = get_log_files(LOG_DIR)

SEPARATOR = '=' * 60
results = {}

for path in files:
    if path.is_file():
        relative = path.relative_to(LOG_DIR)
        results[str(relative)] = read_last_lines(path, NUM_LINES)

print(f'Read {len(results)} log files')

for filename, lines in results.items():
    print(f'FILE: {filename}')
    print(SEPARATOR)
    if lines:
        for line in lines:
            print(line)
    else:
        print('[empty file]')
    print()
```
