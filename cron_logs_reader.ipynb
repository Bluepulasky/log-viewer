# Cron Jobs Log Reader
Reads the first lines of all log files in `/var/log/cron-jobs`

```python
from pathlib import Path
from collections import deque

LOG_DIRS = [
    '/var/log/cron-jobs',
    '/mnt/tank-hdd/config/syncthing/cobian-logs'
]
NUM_LINES = 10

def read_last_lines(filepath, n=NUM_LINES):
    """Read the last n lines of a file."""
    try:
        with open(filepath, 'r', encoding='utf-8', errors='replace') as f:
            lines = [line.rstrip() for line in deque(f, maxlen=n)]
    except PermissionError:
        lines = ['[Permission denied]']
    except Exception as e:
        lines = [f'[Error reading file: {e}]']
    return lines

def get_log_files(log_dir):
    results = {}
    for log_dir_str in LOG_DIRS:
        log_dir = Path(log_dir_str)
        if not log_dir.exists():
            continue
        for path in sorted(log_dir.rglob('*')):
            if path.is_file():
                relative = f"{log_dir.name}/{path.relative_to(log_dir)}"
                lines = read_last_lines(path)
                if lines:
                    results[relative] = lines
    return results

files = get_log_files(LOG_DIRS)

SEPARATOR = '=' * 60
results = {}

for path in files:
    if path.is_file():
        relative = path.relative_to(LOG_DIRS)
        results[str(relative)] = read_last_lines(path, NUM_LINES)

print(f'Read {len(results)} log files')

for filename, lines in results.items():
    print(f'FILE: {filename}')
    print(SEPARATOR)
    if lines:
        for line in lines:
            print(line)
    else:
        print('[empty file]')
    print()
```
